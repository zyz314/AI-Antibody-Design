```python
""""""
"""Assess a betting strategy.

Copyright 2018, Georgia Institute of Technology (Georgia Tech)
Atlanta, Georgia 30332
All Rights Reserved

Template code for CS 4646/7646

Georgia Tech asserts copyright ownership of this template and all derivative
works, including solutions to the projects assigned in this course. Students
and other users of this template code are advised not to share it with others
or to make it available on publicly viewable websites including repositories
such as github and gitlab.  This copyright statement should not be removed
or edited.

We do grant permission to share solutions privately with non-students such
as potential employers. However, sharing with other current or future
students of CS 7646 is prohibited and subject to being investigated as a
GT honor code violation.

-----do not edit anything above this line---

Student Name: Tucker Balch (replace with your name)
GT User ID: tb34 (replace with your User ID)
GT ID: 900897987 (replace with your GT ID)
"""

import numpy as np
import random
import matplotlib.pyplot as plt
import math


def author():
    """
    :return: The GT username of the student
    :rtype: str
    """
    return "yzheng438"  # replace tb34 with your Georgia Tech username.


def gtid():
    """
    :return: The GT ID of the student
    :rtype: int
    """
    return 903649243  # replace with your GT ID number

#实现一个简单的赌博模拟器：
#模拟美国轮盘赌，计算赢/输的概率。
#使用伪代码中的策略进行模拟。


def get_spin_result(win_prob):
    """
    Given a win probability between 0 and 1, the function returns whether the probability will result in a win.

    :param win_prob: The probability of winning
    :type win_prob: float
    :return: The result of the spin.
    :rtype: bool
    """
    spin_result = False
    if np.random.random() <= win_prob:
        spin_result = True
    return spin_result


np.random.seed(gtid())
win_prob = 18 / 38

def experiment1_martingale_simulation(total_spins_time, win_prob, initial_money = 0):
  winnings = np.zeros(total_spins_time +1)# 初始化结果数组
  episode_winnings = initial_money
  bet_amount = 1
  spin_time = 0
  win_prob = 18 / 38
  target_winnings = initial_money + 80

  while spin_time < total_spins_time and episode_winnings < target_winnings:
    won = False
    while spin_time < total_spins_time and episode_winnings < target_winnings and not won:
      spin_time = spin_time + 1
      won = get_spin_result(win_prob)
      if spin_time < total_spins_time and episode_winnings < target_winnings and won:
        episode_winnings = episode_winnings + bet_amount
        bet_amount = 1
      else:
        episode_winnings = episode_winnings - bet_amount
        bet_amount = bet_amount * 2
      winnings[spin_time] = episode_winnings
        # 如果目标收益达到80美元，则填充剩余值
    if episode_winnings >= target_winnings:
        winnings[spin_time:] = target_winnings
  return winnings


#运行模拟实验
# Experiment 1 – Explore the Strategy and Create Charts

# Figure 1: Run 10 episodes, track winnings from 0 each time, and plot all episodes on one chart (X-axis: 0-300, Y-axis: -256 to +100).
def plt_figure_1(experiment_1_results):
    for i in range(10):  # 为每次模拟创建一条线
      plt.plot(experiment_1_results[i, :301])  # 绘制线条
    plt.title('Figure 1: Experiment 1, 10 Simulations Results')  # 添加标题
    plt.xlabel('Spins Times')  # 添加x轴标签
    plt.ylabel('Episode Winning Money')  # 添加y轴标签
    plt.xlim(0, 300)  # 设置x轴的显示范围为0到300
    plt.ylim(-256, 100)  # 设置y轴的显示范围为-256到100
    plt.legend([f'Episode {i + 1}' for i in range(10)])  # 添加图例
    #plt.savefig("images/figure_1.png")  # 先保存图形
    plt.show()  # 显示图形
    plt.clf()  # 清空当前图形，准备绘制下一个

# Figure 2: Run 1000 episodes, plot the mean winnings per spin with standard deviation lines.
def plt_figure_2(experiment_1_results):
    mean = experiment_1_results.mean(axis=0)  # 计算每次旋转的平均赢得金额
    std_dev = experiment_1_results.std(axis=0)  # 计算每次旋转的赢得金额的标准差

    plt.plot(mean, label='Mean',color = 'blue')  # 绘制平均线
    plt.plot(mean + std_dev, label='Mean + Std_Dev', color = 'red')  # 绘制平均值加标准差的线
    plt.plot(mean - std_dev, label='Mean - Std_Dev', color = 'green')  # 绘制平均值减标准差的线
    plt.title('Figure 2: Experiment 1, Mean and Standard Deviation')  # 添加标题

    plt.xlabel('Spins Times')  # 添加x轴标签
    plt.ylabel('Episode Winning Money')  # 添加y轴标签
    plt.xlim(0, 300)
    plt.ylim(-256, 100)
    plt.legend(['Mean', 'Mean + Std_Dev', 'Mean - Std_Dev'])  # 添加图例
    #plt.savefig("images/Figure_2.png")
    plt.show()  # 显示图形
    plt.clf()  # 清空当前图形，准备绘制下一个



# Figure 3: Similar to Figure 2 but plot the median instead of the mean.
    # 运行1000次模拟，这里我们重用了前面的结果
def plt_figure_3(experiment_1_results):
    median = np.median(experiment_1_results, axis=0)  # 计算每次旋转的赢得金额的中位数
    std_dev = experiment_1_results.std(axis=0)  # 计算每次旋转的赢得金额的标准差

    plt.plot(median, label='Median', color = 'blue')  # 绘制中位数线
    plt.plot(median + std_dev, label='Median + Std_Dev', color = 'red')  # 绘制中位数加标准差的线
    plt.plot(median - std_dev, label='Median - Std_Dev', color = 'green',)  # 绘制中位数减标准差的线
    plt.title('Figure 3: Experiment 1 Median and Standard Deviation')  # 添加标题
    plt.xlim(0, 300)
    plt.ylim(-256, 100)
    plt.xlabel('Spins Times')  # 添加x轴标签
    plt.ylabel('Episode Winning Money')  # 添加y轴标签
    plt.legend(['Median', 'Median + Std_Dev', 'Median - Std_Dev'])  # 添加图例
    #plt.savefig("images/Figure_3.png")
    plt.show()  # 显示图形
    plt.clf()  # 清空当前图形，准备绘制下一个


 # Experiment 2 – A More Realistic Gambling Simulator
# Introduce a $256 bankroll limit.\

def experiment2_martingale_simulation(total_spins_time, win_prob, initial_money):
    """
    实现一个带有资金限制的赌博模拟器。

    参数:
    total_spins_time (int): 每次模拟的总旋转次数。
    win_prob (float): 赢得一次赌注的概率，默认为18/38。
    initial_money (int): 初始资金，默认为256美元。

    返回:
    np.ndarray: 一个记录每次旋转后资金状况的数组。
    """
    episode_winnings = 0  # 初始资金为0美元
    winnings = np.zeros(total_spins_time + 1)  # 初始化结果数组
    spin_time = 0  # 初始化旋转次数
    win_prob=18/38
    initial_money=256
    target_winnings = 80

    while episode_winnings < target_winnings and spin_time < total_spins_time and episode_winnings > -initial_money:
        won = False  # 初始化赢得结果为False
        bet_amount = 1  # 初始赌注为1美元

        while not won and spin_time < total_spins_time:
            spin_time = spin_time + 1  # 增加旋转次数
            won = get_spin_result(win_prob)  # 获取旋转结果

            if won:
                episode_winnings = episode_winnings + bet_amount  # 赢得赌注
                if episode_winnings >= target_winnings:
                    episode_winnings = target_winnings  # 目标收益限制为80美元
            else:
                episode_winnings = episode_winnings - bet_amount  # 输掉赌注
                bet_amount = bet_amount * 2
                if bet_amount > initial_money + episode_winnings:
                  bet_amount = initial_money + episode_winnings

            winnings[spin_time] = episode_winnings  # 记录当前收益

    # 如果旋转次数未达上限，填充剩余值为当前资金
    if spin_time < total_spins_time:
        winnings[spin_time:] = episode_winnings

    return winnings


# Figure 4: Run 1000 episodes, plot mean winnings per spin with standard deviation lines.
# 运行1000次模拟


# Figure 4
def plt_figure_4(experiment_2_results):
    mean = np.mean(experiment_2_results, axis=0)
    std_dev = experiment_2_results.std(axis=0)
    mean_plus_sd = mean + std_dev
    mean_minus_sd = mean - std_dev

    plt.plot(mean, color = 'blue', label='Mean')
    plt.plot(mean_plus_sd, color = 'red', label='Mean + Std_Dev')
    plt.plot(mean_minus_sd , color = 'green', label='Mean - Std_Dev')
    plt.title('Figure 4: Experiment 2 Mean and Standard Deviation')
    plt.xlabel('Spins Times')
    plt.ylabel('Episode Winning Money')
    plt.legend(['Mean', 'Mean + Std_Dev', 'Mean - Std_Dev'])
    plt.xlim((0, 300))
    plt.ylim(-256, 100)
    #plt.savefig("images/Figure_4.png")
    plt.show()
    plt.clf()  # 清空当前图形，准备绘制下一个

# Figure 5: Similar to Figure 4 but plot the median instead of the mean.
def plt_figure_5(experiment_2_results):
    median = np.median(experiment_2_results, axis=0)  # 计算每次旋转的赢得金额的中位数
    std_dev = experiment_2_results.std(axis=0)
    median_plus_sd = median + std_dev
    median_minus_sd = median - std_dev
    plt.plot(median, label='Median', color = 'blue')  # 绘制中位数线
    plt.plot(median + std_dev, label='Median + Std_Dev', color = 'red')  # 绘制中位数加标准差的线
    plt.plot(median - std_dev, label='Median - Std_Dev', color = 'green',)
    plt.title('Figure 5: Experiment 2 median and Standard Deviation')
    plt.xlabel('Spins Times')
    plt.ylabel('Episode Winning Money')
    plt.legend(['Median', 'Median + Std_Dev', 'Median - Std_Dev'])
    plt.xlim((0, 300))
    plt.ylim((-256, 100))
    #plt.savefig("images/Figure_5.png")
    plt.show()
    plt.clf()  # 清空当前图形，准备绘制下一个


def test_code():
    """
    Method to test your code
    """
    # Set the win probability to the appropriate value for American roulette
    win_prob = 18 / 38  # 18 out of 38 slots are wins

    # Seed the random number generator with the student's Georgia Tech ID for reproducibility
    np.random.seed(gtid())

    # Run 10 episodes of Experiment 1 and collect the results
    experiment1_results1 = np.array([experiment1_martingale_simulation(1000, win_prob) for _ in range(10)])
    plt_figure_1(experiment1_results1)  # Plot the results of the 10 episodes

    # Run 1000 episodes of Experiment 1 and collect the results
    experiment1_results2 = np.array([experiment1_martingale_simulation(1000, win_prob) for _ in range(1000)])
    plt_figure_2(experiment1_results2)  # Plot the mean and standard deviation of the results
    plt_figure_3(experiment1_results2)  # Plot the median and standard deviation of the results

    # Run 1000 episodes of Experiment 2 and collect the results
    experiment2_results = np.array([experiment2_martingale_simulation(1000, win_prob, 256) for _ in range(1000)])
    plt_figure_4(experiment2_results)  # Plot the mean and standard deviation of the results
    plt_figure_5(experiment2_results)  # Plot the median and standard deviation of the results

    # Analyze the final results of Experiment 1
    exper1_final_results = experiment1_results2[:, -1]  # Get the final winnings for each episode
    times_win_80 = np.sum(exper1_final_results >= 80)  # Count how many times $80 was won
    probability_win_80 = times_win_80 / 1000  # Calculate the probability of winning $80
    print(f"Estimated Probability of Winning $80 in experiment1: {probability_win_80}")

    exper1_expected_value = np.mean(exper1_final_results)  # Calculate the expected value of the winnings
    print(f"Expected result of Winnings after 1000 times spins in experiment1: {exper1_expected_value}")

    # Analyze the final results of Experiment 2
    exper2_final_results = experiment2_results[:, -1]  # Get the final winnings for each episode
    times_win_80 = np.sum(exper2_final_results >= 80)  # Count how many times $80 was won
    probability_win_80 = times_win_80 / 1000  # Calculate the probability of winning $80
    print(f"Estimated Probability of Winning $80 in experiment2: {probability_win_80}")

    exper2_expected_value = np.mean(exper2_final_results)  # Calculate the expected value of the winnings
    print(f"Expected result of Winnings after 1000 times spins in experiment2: {exper2_expected_value}")

    # Calculate the median and standard deviation of the final results in Experiment 2
    exper2_median_result = np.median(exper2_final_results)
    exper2_std_dev_result = np.std(exper2_final_results)
    print(f"Median of Winnings after 1000 times spins in experiment2: {exper2_median_result}")
    print(f"Standard Deviation of Winnings after 1000 times spins in experiment2: {exper2_std_dev_result}")


if __name__ == "__main__":
    test_code()


```
